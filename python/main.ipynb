```python
class Employee:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def get_details(self):
        return f"Hello {self.name} you are {self.age} years old"
    def __str__(self):
        return f"Employee {self.name} is {self.age} years old"
    def __repr__(self):
        return f"Employee {self.name} is {self.age} years old"
alan = Employee("Alan", 25)
print(alan.name)
alan

        
```
